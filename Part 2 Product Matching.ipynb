# Loading and Basic EDA

```python
Amazon = pd.read_csv(r'C:\Users\DELL\Desktop\Assignmet 3\amz_com-ecommerce_sample.csv',encoding='latin-1')
Amazon.head()
```

```python
Flipkart = pd.read_csv(r'C:\Users\DELL\Desktop\Assignmet 3\flipkart_com-ecommerce_sample.csv',encoding='latin-1')
Flipkart.head()
```

# Keeping Required columns form both the datsets


```python
Amazon.drop(columns=['crawl_timestamp', 'product_url','product_category_tree', 'pid',
                 'image', 'is_FK_Advantage_product', 'description', 'product_rating',
                 'overall_rating', 'brand', 'product_specifications'],inplace =True)
```

```python
Flipkart.drop(columns=['crawl_timestamp', 'product_url','product_category_tree', 'pid',
                 'image', 'is_FK_Advantage_product', 'description', 'product_rating',
                 'overall_rating', 'brand', 'product_specifications'],inplace =True)
```

# Renaming columns according to desired output

```python
Amazon.rename(columns={'product_name':'product_Amazon','retail_price':'retail price in Amazon',
                                    'discounted_price':'discounted price in Amazon'},inplace=True)
```

```python
Flipkart.rename(columns={'product_name':'product_Flipkart','retail_price':'retail price in Flipkart',
                                    'discounted_price':'discounted price in Flipkart'},inplace=True)
```

# Merging both the datsets on ( "uniq_id" )


```python
df = pd.merge(Flipkart,Amazon, how='outer')
df.head()
```

```python
#dropping "uniq_id" column 

df.drop('uniq_id',axis=1,inplace=True)
```

# Module function to return the desired output


```python
def new_data():
    x = input("Enter Required Product name: ")
    
    #returns matched output
    return df[(df['product_Flipkart'] == x) | (df['product_Amazon'] == x)]
```

```python
Amaz_flikt = new_data()
Amaz_flikt
```
